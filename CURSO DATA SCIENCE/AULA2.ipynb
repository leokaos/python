```python
#Notas do(a) estudante
notas = {'19 Trimestre': 8.5, '2° Trimestre': 9.5, '3º trimestre': 7}
notas
```

```python
# Calculando a soma
soma = 0

for nota in notas.values():
    soma += nota

soma
```

```python
# Usando a função embutida sum()
somatorio = sum(notas.values())
somatorio
```

```python
# Usando a função embutida len()
qtd_notas = len(notas)
qtd_notas
```

```python
# Calculando a média
media = round(somatorio / qtd_notas, 1)
media
```

```python
def media(notas):
    valor_media = sum(notas) / len(notas)
    return valor_media

# Notas do(a) estudante
notas = [8.5, 9.0, 6.0, 10.0]
media(notas)
```
