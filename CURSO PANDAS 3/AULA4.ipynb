```python
import pandas as pd
import numpy as np

dados = pd.json_normalize(pd.read_json('dados_hospedagem.json')["info_moveis"])

colunas = list(dados.columns)

dados = dados.explode(colunas[3:])
dados.reset_index(inplace=True, drop=True)

col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas', 'max_hospedes']

dados[col_numericas] = dados[col_numericas].astype(np.int64)

dados[['taxa_deposito','taxa_limpeza','preco', 'avaliacao_geral']] = dados[['taxa_deposito','taxa_limpeza','preco', 'avaliacao_geral']].map(lambda x: x.replace('$','').replace(',', '').strip()).astype(np.float64)

dados['descricao_local'] = dados['descricao_local'].str.lower().replace('[^a-zA-Z0-9\-\']',' ', regex=True).replace('(?<!\w)-(?!\w).',' ', regex=True)
dados['descricao_local'] = dados['descricao_local'].str.split(" ")

dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
dados['comodidades'] = dados['comodidades'].str.split(',')

dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower().replace('[^a-zA-Z0-9\-\']', ' ', regex=True).replace('(?<!\w)-(?!\w)', '', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split(" ")
```
