```python
import pandas as pd
import numpy as np

dados = pd.json_normalize(pd.read_json("dados_locacao_imoveis.json")["dados_locacao"])

dados = dados.explode(list(dados.columns)[1:])
dados.reset_index(drop=True,inplace=True)

dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x: x.replace('$','').replace('reais', '').replace(',', '.').strip()).astype(np.float64)
```

```python
dados['apartamento'] = dados['apartamento'].str.replace("(blocoAP)","")
```
