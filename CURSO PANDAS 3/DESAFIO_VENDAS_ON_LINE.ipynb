```python
import pandas as pd
import numpy as np

dados = pd.json_normalize(pd.read_json("dados_vendas_clientes.json")["dados_vendas"])
```

```python
dados = dados.explode(['Cliente', 'Valor da compra'])
dados.reset_index(drop=True,inplace=True)

dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','').strip()).astype(np.float64)
```

```python
dados['Cliente'] = dados['Cliente'].str.lower().replace('[^a-z ]','', regex=True)
```
