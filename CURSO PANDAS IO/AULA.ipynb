# AULA 5

Nesta aula, vamos utilizar duas bibliotecas: pandas e SQLAlchemy. É importante ter certeza de que a versão do SQLAlchemy que você está utilizando seja compatível com a biblioteca pandas, para que os códigos da aula possam ser executados corretamente.

```python
import pandas as pd
import sqlalchemy

sqlalchemy.__version__
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
dados = pd.read_csv('datasets/clientes_banco.csv')
```

```python
dados.head()
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

```python
empregados.to_sql('empregados', con=engine, index=False)
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
query = 'SELECT * FROM clientes'
pd.read_sql(query, engine) 
```

```python
query = text('DELETE FROM clientes WHERE ID_Cliente=5008804')

with engine.connect() as conn:
    conn.execute(query)
```
