```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
dados.query('@imoveis_comerciais not in Tipo')
```

```python
df = dados.query('@imoveis_comerciais not in Tipo')
df.head()
```

```python
df.Tipo.unique()
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14,10), color='purple')
```

```python
df.Tipo.unique()
```

```python
df_percetual_imoveis = df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')

df_percetual_imoveis.plot(kind='bar', figsize=(10, 14), color='green', xlabel='Tipos', ylabel='Percentual')
```

```python
df = df.fillna(0)
```

```python
registros_remover = df.query("Valor == 0 | Condominio == 0").index
```

```python
df.drop(index=registros_remover, axis=0, inplace=True)
```

```python
df.query("Valor == 0 | Condominio == 0")
```

```python
df.drop('Tipo', axis=1, inplace=True)
```

```python
df['Quartos'] == 1
```

```python
apartamentos_1_quarto_aluguel_menor_1200 = df[ (df['Quartos'] == 1) & (df['Valor'] < 1200) ]
apartamentos_1_quarto_aluguel_menor_1200
```

```python
apartamentos_pelo_menos_2_quartos_aluguel_menor_3000_area_maior_70 = df[ (df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70) ]
apartamentos_pelo_menos_2_quartos_aluguel_menor_3000_area_maior_70
```

```python
df.to_csv('dados_apartamento.csv', index=False, sep=';')
```
