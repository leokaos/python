```python
import pandas as pd

dados = pd.read_csv('arq_municipios_fronteiricos.csv')
```

```python
def remove_numeros_hifens(nome):
    nome_tratato = ''.join( [ i for i in nome if not i.isdigit() ] )
    return nome_tratato.replace('–', '').replace('-', '').strip()
```

```python
dados.loc[:,'Município'] = dados['Município'].apply(remove_numeros_hifens)
dados
```
