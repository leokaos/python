```python
import requests
```

```python
url = "http://localhost:5000/cotacao"
```

```python
dados = {
    "tamanho": 120,
    "garagem": 2,
    "ano": 2001
}
```

```python
auth = requests.auth.HTTPBasicAuth('leo', 'leo12345')
```

```python
response = requests.post(url, json=dados, auth=auth)
```
